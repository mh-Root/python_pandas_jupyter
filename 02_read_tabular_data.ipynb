__Načtení dat do pandas (csv, excel, sql)__

```python
import pandas as pd
```

metoda DataFrame.read_excel(), read_csv(), read_table() načte data ze souboru

```python
# parametr 'usecols' umožní výběr jednotlivých sloupců
apple = pd.read_excel('../data/AppleStore.xlsx', usecols = ["id","track_name","size_bytes","currency","price","rating_count_tot","rating_count_ver","user_rating","ver","cont_rating","prime_genre"])
```

funkce head()<br>
funkce head() zobrazí začátek datového souboru, implicitně 5 řádků, parametr lze měnit (v příkladu prvních 10 řádků)

```python
funkce head() zobrazí začátek datového souboru, implicitně 5 řádků, parametr lze měnit (v příkladu prvních 10 řádků)
apple.head(10)
```

funkce tail()<br>
funkce head() zobrazí konec datového souboru, implicitně 5 řádků, parametr lze měnit jako v případě funkce head()

```python
apple.tail()
```
