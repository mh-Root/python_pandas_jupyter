__Import dat bez výběru indexu__

```python
df2 = pd.read_csv('../data/drinks.csv')
```

```python
df2.head()
```

Pokud není určen sloupec, který bude indexem při prvním čtení dat, použije pandas jako index celá čísla 0 až n-1. Tím se technicky vytvoří objekt __RangeIndex__.

```python
df2.index
```

Je to podobné jako v jazyce Python objekt 'range'

```python
range(7)
```

Konverze těchto objektů do seznamu produkuje stejnou věc:

```python
list(df2.index)
```

```python
list(range(7))
```

Prozatím není vůbec důležité mít RangeIndex. Výběr z toho se děje stejně jako u .loc a .iloc. Některé příklady:

```python
df2.loc[[2, 4, 5], ['beer', 'wine']]
```

```python
df2.iloc[[2, 4, 5], [3,2]]
```

```python
df2.loc[:3]
```

```python
df2.iloc[:3]
```

__Výběr sloupce DataFrame s tečkovou notací__

Pandas umožňuje vybrat jeden sloupec jako objekt typu Series pomocí tečkového zápisu. Toto se také nazývá přístup k atributům. Jednoduše lze umístit název sloupce bez uvozovek za tečku a DataFrame takto:

```python
df2.country
```

```python
df.continent
```

Výhodou výběru sloupců, jako je tato, je to, že lze získat pomoc při zřetězeních metod po výběru. Pokud se např. za název sloupce umístí další tečka, stiskem klávesy __TAB__, se v rozbalovací nabídce zobrazí seznam všech metod série.
