__Výrazy pro více podmínek: "&" = a, "|" = nebo, "~" = ne__

```python
criteria_1 = so['score'] >= 5
criteria_2 = so['ans_name'] == 'Mike Pennington'
criteria_2
criteria_1
```

__Kombinace kritérií__

```python
criteria_all = criteria_1 & criteria_2
```

```python
so[criteria_all]
```

```python
criteria_2.head()
```

Závorky k oddělení podmínek

```python
(so['score'] >= 5) & (so['ans_name'] == 'Mike Pennington')
```

Stejné jako předchozí

```python
so[(so['score'] >= 5) & (so['ans_name'] == 'Mike Pennington')]
```

Použití operátoru __or__

```python
so[(so['score'] >= 100) | (so['answercount'] >= 10)].head()
```

Použití operátoru ~

ne skore menší než '100'

```python

so[~ (so['score'] <= 100)].head()
```

```python
~ ((so['score'] >= 100) | (so['answercount'] >= 10))
```

__Komplexní podmínky__

```python
criteria_1 = (so['score'] >= 5) & (so['ans_name'] == 'Nick Crawford')
criteria_2 = (so['answercount'] >=5) & (so['ans_name'] == 'Ted Petrou')
criteria_all = criteria_1 | criteria_2
so[criteria_all]
```

__Metoda isin()__

```python
criteria = so['ans_name'].isin(['Nick Crawford','Ted Petrou','unutbu'])
```

```python
so[criteria].head()
```
