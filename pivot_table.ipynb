### knihovna pandas - kontingenční tabulka 1

```python
import pandas as pd
```

```python
drinks = pd.read_csv('../data/drinks.csv')
```

```python
drinks.columns = ['Stát','Pivo','Lihoviny','Víno','Celkem','Kontinent']
```

```python
drinks.head()
```

```python
drinks.groupby('Kontinent').mean()
```

```python
drinks.pivot_table(index=['Kontinent','Stát'])
```

```python
drinks.pivot_table(index=['Kontinent'], values=['Pivo','Víno','Lihoviny'],aggfunc=['mean'])
```

```python
drinks.pivot_table(index=['Kontinent'], values=['Pivo','Víno','Lihoviny'],aggfunc=['min','max'])
```
