<font color=blue>__Třídění a výběr záznamů__</font>

```python
import pandas as pd
```

<font color=red>parametr usecols</font>

```python
# parametr 'usecols' umožní výběr jednotlivých sloupců
imdb = pd.read_csv('../data/imdb_1000.csv', usecols = ['star_rating','title','content_rating','genre','duration'])
```

```python
imdb.head()
```

<font color=red>metoda sort_values - Series</font>

```python
# sort_values - třídění (pandas Series) sloupce podle abecedy (vzestupně)
imdb.title.sort_values().head()
```

```python
# sort_values - třídění sloupce podle abecedy (sestupně -> ascending = False)
imdb.title.sort_values(ascending = False).head()
```

<font color=red>metoda sort_values - DataFrame</font>

```python
imdb.sort_values('title').head()
```

<font color=red>metoda sort_values - DataFrame (třídění podle několika sloupců)</font>

```python
imdb.sort_values(['title','duration'], ascending=False).head()
```

```python
imdb.sort_values(['genre','title']).head(20)
```

<font color=red>třídění pomocí logických operátorů 1 (podmínka pro výběr 'a', značí se znakem '&')</font>

```python
# výběr záznamů, které mají hodnotu větší než ... a zároveň patří do ...
imdb[(imdb.star_rating >= 6.5) & (imdb.genre == 'Horror')].head(15)
```

<font color=red>třídění pomocí logických operátorů  (podmínka pro výběr 'nebo', značí se znakem '|')</font>

```python
# výběr záznamů, které patří do ... nebo do ... nebo do ... 
imdb[(imdb.genre == 'Action') | (imdb.genre == 'Drama') | (imdb.genre == 'Crime')].head(15)
```

<font color=red>metoda isin() - jednodušší metoda pro výběr</font>

```python
imdb[imdb.genre.isin(['Action','Drama','Crime'])].head(15)
```
